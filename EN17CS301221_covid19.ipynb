```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('/content/nation_level_daily.xlsx')
```

```python
df.head()
```

```python
plt.scatter(x=df['date'],y=df['totalconfirmed'])
plt.show()
```

```python
df.date=pd.to_datetime(df.date)
df.set_index('date', inplace=True)
```

```python
df.head()
```

```python
df.plot(kind='bar',subplots=True,layout=(6,2) ,sharex=True ,sharey=False, figsize=(15,20))
plt.show()
```

```python
df.plot(kind='line',subplots=True,layout=(6,2) ,sharex=True ,sharey=False, figsize=(15,20))
plt.show()
```
